# ex1) $y = x^{2}$

컴퓨터에서 $x^{2}$ 그래프의 최솟값 찾기

- $f(x) = x^{2}$
- $\frac{dy}{dx} = 2x$
- $x_{new} = x_{old} - \alpha * 2x_{old}$ <br>

$x_{0} = 1,\alpha = 0.1$일 때 x의 변화를 살펴보겠습니다.

|<div align="center">n step|<div align="center"> Value                        |
| ------------- |:-------------------------------------------------------------------:
|<div align="center">0 |<div align="left">$x_{0}$ = 1                              |
|<div align="center">1 |<div align="left">$x_{1} = 1 - 0.1 \times 2 \times 1 = 0.8$|  
|<div align="center">2 |<div align="left">$x_{2} = 0.8 - 0.1 \times 2 \times 0.8 = 0.64$|
|<div align="center">3 |<div align="left">$x_{3} = 0.64 - 0.1 \times 2 \times 0.64 = 0.512$|
|<div align="center"> ... |<div align="left">... |

Gradient Descent 에서는 2개의 hyper-parameter가 존재한다.
1. learning rate
2. epoch

learning rate가 너무 작으면 최소점 도달에 너무 오래걸리거나 도달하지 못합니다. 반면 너무 큰 값을 가지면 발산하게 됩니다.

[Gradient Descent 실습](http://onmyphd.com/?p=gradient.descent)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10, 1)
f_x = x ** 2

plt.plot(x, f_x)
plt.show()
```

```python
x_new = 10
gd = []
y = []
learning_rate = 0.1
for i in range(100):
    old_value = x_new
    x_new = old_value - learning_rate * 2 * old_value
    gd.append(x_new)
    y.append(x_new ** 2)    
```

```python
plt.plot(x, f_x)
plt.scatter(gd, y)
plt.show()
```

## ex2) $y = x sin(x^{2}) + 1$

[울프람 알파](http://www.wolframalpha.com/) 이용하여 미분을 계산해 본다.

- $ f(x) =  x sin(x^{2}) + 1 $ <br>
- $ \frac{df(x)}{dx} = sin(x^{2}) + 2x^{2}cos(x^{2})$
- $ x := x - \alpha \times (sin(x^{2}) + 2x^{2}cos(x^{2})) $


```python
def f(x):
    return x * np.sin(x ** 2) + 1

def d_f(x):
    return np.sin(x**2) + 2 * (x ** 2) * np.cos(x ** 2)
```

```python
x = np.arange(-3, 3, 0.001)
f_x = f(x)
```

```python
plt.plot(x, f_x)
plt.show()
```

```python
x_new = 1
gd = []
y = []
learning_rate = 0.01

for i in range(10000):
    old_value = x_new
    x_new = old_value - learning_rate * d_f(old_value)
    gd.append(x_new)
    y.append(f(x_new))
    
plt.plot(x, f_x)
plt.scatter(gd, y)
plt.show()
```
